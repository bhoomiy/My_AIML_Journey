```python
import pandas as pd
```

```python
df=pd.read_csv("IRIS.csv")
df.head(10)
print(df.shape)
print(df.dtypes)
df.isna().sum()
df.describe()
```

```python
df[(df["petal_length"]>4.5) & (df["species"]=="Iris-virginica")]
```

```python
df.groupby("species").agg(
    avg_sepal_length=("sepal_length","mean"),
    max_petal_width=("petal_width","max"),
    std_sepal_width=("sepal_width","std")
)
```

```python
df["petal_ratio"]=df["petal_length"]/df["petal_width"]
df.groupby("species")["petal_ratio"].mean()
```

```python
df1=pd.read_csv("Titanic-Dataset.csv")
df1
```

```python
df1[["Name","Sex","Age","Fare","Survived"]]
```

```python
df1[(df1["Sex"]=="female") & (df1["Fare"]>30)]
```

```python
df1.groupby(["Pclass","Survived"]).agg(
        sr=("Survived","count"),
        avg_fare=("Fare","mean"),
        avg_age=("Age","mean")
)
```

```python
df1.groupby("Pclass").agg(
    {
        "Survived":"mean"
    }
)
```

```python
df=pd.read_csv("movies.csv")
df.isna().sum()
#remove duplicates
cleaned=df.copy()
cleaned=cleaned.drop_duplicates()
#handle missing values
cleaned["Gross"]=cleaned["Gross"].fillna(0)
cleaned["RATING"]=cleaned["RATING"].fillna(0.0)
cleaned["RATING"]=cleaned["RATING"].astype("int64")
cleaned.fillna(0)
```
