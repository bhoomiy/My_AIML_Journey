```python
import gymnasium as gym
import numpy as np
import random
```

```python
gamma=0.99
alpha=0.5
epsilon=0.1
episodes=500
```

```python
#Q table->stores Q values
Q=np.zeros((48,4))
```

```python
env=gym.make("CliffWalking-v1")
```

```python
#policy-->epsilon greedy
def epsilon_greedy(state):
    if random.random()<epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(Q[state])
```

```python
for episode in  range(episodes):
    render=(episode%50==0)
    if render:
        env=gym.make("CliffWalking-v1",render_mode="human")
    else:
        env=gym.make("CliffWalking-v1")

    done=False
    state,_=env.reset()
    episode_len=0
    total_reward=0
    while not done:
        action=epsilon_greedy(state)
        next_state,rewards,terminated,truncated,_=env.step(action)
        done=terminated or truncated

        #Q-learning
        Q[state,action]+=alpha * (rewards + gamma  * np.max(Q[next_state])- Q[state,action])
        state=next_state
        episode_len+=1
        total_reward+=rewards

    print(f"for episode={episode+1}: total reward={total_reward} and episode length={episode_len}")
    env.close()
```

```python
#what did our agent learn
env=gym.make("CliffWalking-v1",render_mode="human")
state,_=env.reset()
total_reward=0
episode_len=0
done=False
while not done:
    action=np.argmax(Q[state])
    state,reward,terminated,truncated,_=env.step(action)
    done=terminated or truncated
    total_reward+=reward
    episode_len+=1
print(f"for episode={episode+1}: total reward={total_reward} and episode length={episode_len}")
env.close()
```
