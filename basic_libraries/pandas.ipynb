```python
#Missing values
df=pd.read_csv("raw_data.csv")
df.isna().sum()
cleaned_data=df.copy()
avg_mean=cleaned_data["age"].mean()
cleaned_data["age"]=cleaned_data["age"].fillna(avg_mean)

avg_mean=cleaned_data["income"].mean()
cleaned_data["income"]=cleaned_data["income"].fillna(avg_mean)
cleaned_data
```

```python
#Handle data types
df.dtypes
df2=df.copy()
df2=df2.fillna(0)
df2["age"]=df2["age"].astype("int64").copy()
df2.dtypes
```

```python
df2
```

```python
#Feature Engineering
df2["tax"]=df2["income"].apply(lambda x: "20%" if x>50000 else "10%")

gender_map={"Male": "M","Female":"F","Unknown":"U"}
df2["gender"]=df2["gender"].map(gender_map)

df2=df2.assign(new_income=df2["income"]*1.1)
df2
```

```python
df2.columns=["name","age","country","gender","income","tax","new_income","id"]
df2
```

```python
#write to csv
df2=df.copy()
df2=df2.drop_duplicates()
df2=df2.fillna(0)
df2=df2.sort_values("income")
df2=df2.reset_index()
df2.to_csv("sorted_income.csv")
```
