```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("data.json")
```

```python
#clean and structuring the data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    clean_data=[]
    unique_users=set()
    for user in data:
        #clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating

        #handle missing age values
        raw_age=user.get("age")
        if (raw_age==None):
            user["age"]=None

        #duplicate data
        if (user["name"] in unique_users):
            continue

        unique_users.add(user["name"])
        clean_data.append(user)

    return clean_data
```

```python
cleaned_data=clean_data(data)
```

```python
def get_insights(data):
    #avg rating
    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])
    print(f"Avg rating is {tot_rating/len(data)}")

    #percentage of users with poor rating
    poor_rating=0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+=1
    print(f"Percentage of users with poor rating is {poor_rating/len(data) *100}%")
```

```python
get_insights(cleaned_data)
```
