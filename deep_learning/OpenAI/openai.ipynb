```python
#AI personal assistant
#queries->answer
#emails->summarize

class PersonalAssistant:
    def __init__(self):
        print("Hiee!!! I'm your personal assistant. How can I help you??")

    def answer_query(self):
        question=input("How can I help you??")
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",  # or another Groq-supported model
            messages=[
                {"role":"user","content":question},
                {"role":"system","content":"act like a helpful assistant"}
            ],
            temperature=0.7,
            max_completion_tokens=512
        )

        print(response.choices[0].message.content.strip())

    def summarize_email(self):
        email_text=input("Please text your email body message: ")
        prompt=f"Summarize the following email in 2-3 sentences: {email_text}"
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",  # or another Groq-supported model
            messages=[
                {"role":"user","content":email_text},
                {"role":"system","content":"act like an expert email assistant"}
            ],
            temperature=0.3,
            max_completion_tokens=512
        )

        print("Summary: ",response.choices[0].message.content.strip())
```

```python
assistant=PersonalAssistant()
```

```python
assistant.answer_query()
```

```python
assistant.summarize_email()
```

```python
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO
import os
from dotenv import load_dotenv

load_dotenv()

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

response = client.models.generate_content(
    model="gemini-2.5-flash-image",
    contents="A cute baby dragon sleeping on a pile of books, digital art",
    config=types.GenerateContentConfig(
        response_modalities=["TEXT", "IMAGE"]
    )
)

for part in response.candidates[0].content.parts:
    if part.inline_data:
        image = Image.open(BytesIO(part.inline_data.data))
        image.save("dragon.png")
        image.show()
```
