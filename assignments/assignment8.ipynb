```python
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
#Q1
import numpy as np
scores=np.random.normal(70,10,200) #normal disribution
plt.hist(scores,bins=20,edgecolor="black")
m=scores.mean()
plt.axvline(x=m,linestyle="--",color="red")
```

```python
#Q2
penguins=sns.load_dataset("penguins")
sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)
```

```python
#Q3
tips=sns.load_dataset("tips")
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
```

```python
#Q4
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
```

```python
#Q5
average_tip=tips.pivot_table(index="day",columns="time",values="tip",aggfunc="mean")
sns.heatmap(
    data=average_tip,
    cmap="coolwarm",
    annot=True,
    fmt=".2f"
)
```
