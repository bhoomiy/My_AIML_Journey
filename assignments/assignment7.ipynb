```python
import requests as req
from bs4 import BeautifulSoup
page_count=1

while True:
    URL=f"https://quotes.toscrape.com/page/{page_count}/"
    res=req.get(URL)

    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")

    if not quotes:
        print("No content anymore")
        break

    with open(f"scraped_data/quotes{page_count}.html","w") as f:
        f.write(res.text)
        print(f"Downloaded page{page_count}")

    page_count+=1
```

```python
#Extract useful info-page 1
with open("scraped_data/quotes1.html","r") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content,"lxml")
```

```python
all_quotes=soup.select("div.quote")
life_quotes=[]
for q in all_quotes:
    all_tags=[]

    for t in q.select(".tags .tag"):
        all_tags.append(t.get_text())
    
    if "life" in all_tags:
        text=q.select_one("span.text").get_text()
        author=q.select_one("small.author").get_text()

        life_quotes.append([text,author])
```

```python
#Extract for all
life_quotes=[]
for page_count in range(1,11):
    with open(f"scraped_data/quotes{page_count}.html","r") as f:
        html_content=f.read()
    soup=BeautifulSoup(html_content,"lxml")
    all_quotes=soup.select("div.quote")
    for q in all_quotes:
        all_tags=[]

        for t in q.select(".tags .tag"):
            all_tags.append(t.get_text())
        
        if "life" in all_tags:
            text=q.select_one("span.text").get_text()
            author=q.select_one("small.author").get_text()

            life_quotes.append([text,author])
```

```python
import pandas as pd

df=pd.DataFrame(life_quotes,columns=["Quote","Author"])
df.to_csv("cleaned_data/lifequotes.csv",index=0)
```
